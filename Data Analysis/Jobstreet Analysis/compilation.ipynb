```python
#Luzon 
filter_luzon  = (merge["Location"] != "Central Visayas") & (merge["Location"] != "Armm") & (merge["Location"] != "Calabarzon & Mimaropa") & (merge["Location"] != "Cebu (Cebu City)") & (merge["Location"] != "Davao") & (merge["Location"] != "Cebu (Others)") & (merge["Location"] != "Philippines") & (~merge["Location"].str.contains("Iloilo")) & (~merge["Location"].str.contains("Davao")) & (~merge["Location"].str.contains("del")) & (~merge["Location"].str.contains("Negros")) & (~merge["Location"].str.contains("Albay")) & (~merge["Location"].str.contains("Bukidnon")) &(~merge["Location"].str.contains("New")) & (~merge["Location"].str.contains("Bahrain")) & (~merge["Location"].str.contains("Saudi")) & (~merge["Location"].str.contains("Caraga")) & (~merge["Location"].str.contains("Palawan")) & (~merge["Location"].str.contains("Bicol"))
count_jobs_luzon = merge.loc[filter_luzon]["Location"].value_counts()
plt.figure(figsize=(20, 5))
plt.bar(count_jobs_luzon.index, count_jobs_luzon.values)
plt.xticks(rotation="30", ha="right")
plt.title('Luzon')
plt.show()
```

```python
# Salary Distribution
salary_append = []
df = merge.loc[merge["Salary"] != "None"]
df["Salary"] = df["Salary"].str.replace(",", "")
df["Salary"] = df["Salary"].str.replace("PHP", "")
df["Salary"] = df["Salary"].str.replace("-", "")
df["Salary"] = df["Salary"].str.split("  ")



salary_df = df["Salary"].reset_index()
salary_df = salary_df.drop(columns="index")

for i in range(len(salary_df)):
    salary = salary_df["Salary"][i]
    for j in range(len(salary)):
        salary_append.append(int(salary[j]))
```

```python
df_salary = pd.DataFrame(salary_append, columns=["Salary"])
# bins = [10,10000, 20000,30000,40000,50000,60000,70000,80000,90000,100000,110000,120000,130000,140000,150000,160000,170000,180000,190000,200000,210000,220000]
bins = [10000,20000,30000,40000,50000,60000,70000,80000,90000,100000,110000,120000,130000,140000,150000,160000,170000,180000,190000,200000,210000,220000]
plt.figure(figsize=(20,10))
plt.hist(df_salary["Salary"],bins=bins, edgecolor="black")
plt.axvline(df_salary["Salary"].median(), linewidth=5,color="#fc4f30", label="Median Salary is PHP {}".format(df_salary["Salary"].median()))
plt.axvline(df_salary["Salary"].mean(), linewidth=5,color="orange", label="Mean Salary is PHP {}".format(df_salary["Salary"].mean()))
plt.legend()
```
