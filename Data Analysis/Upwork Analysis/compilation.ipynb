```python
# skill 
merge['Skill'] = merge["Skill"].str.replace("'","")
merge['Skill'] = merge["Skill"].str.split(",")
```

```python
skill_append = []
for skill in range(len(merge['Skill'])):
    skills = merge['Skill'][skill]
    for k in range(len(skills)):
        skill_append.append(str(skills[k]).strip())
```

```python
merge["Combine Skills"] = merge["Skill_one"] + "," + merge["Skill_two"] + "," + merge["skill_three"] + "," + merge["skill_four"]
```

```python
merge["Combine Skills"] = merge["Combine Skills"].str.split(",")
```

```python
df = merge.loc[~merge["Combine Skills"].isna(),"Combine Skills"]
```

```python
df = df.reset_index().drop(columns=["index"])
```

```python
other_skill_append = []
for i in range(len(df["Combine Skills"])):
    count = df["Combine Skills"][i]
    for j in range(len(count)):
        other_skill_append.append(count[j])
```

```python
skill_without_space = []
filter_skill = list(filter(None, other_skill_append))
for i in filter_skill:
    if i != " ":
        skill_without_space.append(i)
```

```python
final_skill = skill_without_space + skill_append
len(final_skill)
```

```python
skill_frame = pd.DataFrame(final_skill, columns=["Skill"])
count_skill = skill_frame["Skill"].value_counts()[0:20]
sort = count_skill.sort_values()
plt.figure(figsize=(10,5))
plt.title("Top 20 Skills for all Creative Works")
plt.barh(sort.index,sort.values)
plt.show()
```

```python
merge["Est.Proj-Budget"] = merge["Est.Proj-Budget"].str.replace("$","")
merge["Est.Proj-Budget"] = merge["Est.Proj-Budget"].str.replace(",","")
merge["Est.Proj-Budget"] = merge["Est.Proj-Budget"].str.strip()
df = merge.loc[merge["Est.Proj-Budget"] != "None","Est.Proj-Budget"]
```

```python
budget_append = []
budget = df.to_frame()
rest_budget = budget["Est.Proj-Budget"].reset_index()
rest_budget = rest_budget.drop(columns="index")
final_budget = rest_budget.loc[~rest_budget["Est.Proj-Budget"].isna()]

final_budget = final_budget.reset_index()

for i in range(len(final_budget)):
    budget_append.append(final_budget["Est.Proj-Budget"][i])
```

```python
budget_append = [int(i) for i in budget_append]
bins = [0,100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000]
plt.figure(figsize=(20,10))
plt.hist(budget_append,bins=bins,edgecolor='black')
plt.ylabel("Number of Estimated Project Budget")
plt.xlabel("Project Budget Estimation")
plt.title("Distribution of Project Budget Estimation")
plt.axvline(final_budget["Est.Proj-Budget"].astype(int).median(), linewidth=5,color="#fc4f30", label="Median Salary is PHP {}".format(final_budget["Est.Proj-Budget"].astype(int).median()))
plt.axvline(final_budget["Est.Proj-Budget"].astype(int).mean(), linewidth=5,color="black", label="Mean Salary is PHP {}".format(round(final_budget["Est.Proj-Budget"].astype(int).mean()),2))
plt.legend()
```
