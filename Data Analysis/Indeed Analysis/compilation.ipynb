```python
# Remote Jobs
filter = (merge["Location"].str.contains("Remote")) & (merge["Location"] != "Philippines") & (merge["Location"] != "Remote") & (merge["Location"] != "Remote in Philippines") &  (merge["Location"] != "Remote in National Capital Region")
count_jobs = merge.loc[filter]
top_10 = count_jobs["Location"].value_counts()[0:10]
plt.pie(top_10.values, labels = top_10.index,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Remote Job Location")
plt.rcParams['figure.facecolor'] = 'white'
plt.show()
```

```python
# Not Remote Jobs
filter = (~merge["Location"].str.contains("Remote")) & (merge["Location"] != "Philippines")
count_jobs = merge.loc[filter]
top_10 = count_jobs["Location"].value_counts()[0:10]
plt.pie(top_10.values, labels = top_10.index,autopct='%1.1f%%')
plt.title("Not Remote Job Location")
plt.rcParams['figure.facecolor'] = 'white'
plt.axis('equal')
plt.show()
```

```python
# Overall  Location
filter = (merge["Location"] != "Philippines") & (merge["Location"] != "Remote") & (merge["Location"] != "Remote in Philippines") &  (merge["Location"] != "Remote in National Capital Region")
count_jobs = merge.loc[filter]
top_10 = count_jobs["Location"].value_counts()[0:10]
plt.pie(top_10.values, labels = top_10.index,autopct='%1.1f%%')
plt.title("Overall Job Location")
plt.rcParams['figure.facecolor'] = 'white'
plt.axis('equal')
plt.show()
```

```python
# Overall Company
count_company = merge["Company"].value_counts()[0:10]
sort = count_company.sort_values()
plt.barh(sort.index, sort.values)
plt.title("Top 10 Companies")
plt.show()
```

```python
# Non Remote Company
filter = (~merge["Location"].str.contains("Remote")) & (merge["Location"] != "Philippines")
filter_company = merge.loc[filter]
count_company = filter_company["Company"].value_counts()[0:10]
sort = count_company.sort_values()
plt.barh(sort.index, sort.values)
plt.title("Top 10 Companies ( Not Remote )")
plt.show()
```
