```python
def generate_url(job_title):
    url_template = "https://www.jobstreet.com.ph/en/job-search/{}-jobs/"
    format_url = url_template.format(job_title)
    return format_url
```

```python
url = generate_url("Game Designer")
```

```python
res = requests.get(url,headers=headers)
res
```

```python
soup = bs(res.text, "html.parser")
soup.prettify()
```
